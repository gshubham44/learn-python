## 7.12 Subplot
So far, we have used subplot having same width and height. The situation might arise, when we need to increase the size for some subplot. In the following section, we will try to plot such in such case. Firs, we will use *plt.subplot*, itself to make some particular subplot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(25)
y = np.arccos(x)

plt.subplot(221)
plt.scatter(x,y)

plt.subplot(223)
plt.scatter(x,y)

plt.subplot(122)
plt.scatter(x,y)
plt.tight_layout()
plt.show()
```

Above figure shows the resulted plot. The plt.tight_layout() increase the readability of the ticks labels. If this option is not used, then you might have got figures with overlapping labels etc. This options prevents overlapping of axis, title etc.

Now, we will make these kinds of subplot using subplot2grid. Created plot is shown below.

```python
fig = plt.figure()
fig.subplots_adjust(wspace=0.5, hspace=0.4)

ax1 = plt.subplot2grid((3, 3), (0, 0))
ax2 = plt.subplot2grid((3, 3), (0, 1), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 0), colspan=2, rowspan=2)
ax4 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)

ax1.scatter(10*x,y)
ax2.scatter(10*x,y)
ax3.scatter(10*x,y)
ax4.scatter(10*x,y)
plt.show()
```
