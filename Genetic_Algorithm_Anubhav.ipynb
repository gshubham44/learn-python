# Genetic Algorithm
### Introduction
Genetic Algorithm (GA) are search based optimization algorithms inspired by Darwin theory of natural selection i.e., based on the concepts of biological evolution, natural selection, and genetics. It reflects the process of selection of fittest element naturally, where the fittest element dominates over the weaker ones. A GA starts with initial population, from the initial population, this algorithm produces a new population by mimicking the biological mechanisms of evolution, such as selection, crossover, and mutation steps. 

The fitness function helps the algorithm to generate optimal or near optimal solution. The algorithm continues and evolves the population through selection, crossover, and mutation operations. It generates several populations until it satisfies the optimization constraints. GA is frequently used to find optimal or near optimal solutions to difficult problems which otherwise would take a lifetime to solve. It is frequently used to solve optimization problems, in research and machine learning. 

### Introduction to Optimization
GA’s were developed by John Holland and his students and colleagues at the University of Michigan and has since been tried on various optimization problems with high degree of success. Optimization is process of making something better. In any process, we have inputs and set of outputs. The optimization refers to finding the values of inputs in such a way that we get the “best” output values. The definition of best varies from problem to problem, but in mathematical terms it refers maximizing and minimizing the objective function, by varying input parameters.
The set of all possible solutions or values which the inputs can take make up the search space. In this search space, lies a point or set of points which gives optimal solution. The aim of optimization is to find that point or set of points in the search space.

### Genetic Algorithm (GA)

Nature gas always been a great source of inspiration to all mankind. GA’s are search based algorithms based on concepts of natural selection and genetics. GA’s are a subset of much larger branch of computation known as Evolutionary Computation. In GA’s we have population of possible solutions to the given problem. These solutions then undergo recombination and mutation (like in natural genetics), producing new children and the process is repeated over various generations. Each individual (candidate solution) is assigned a fitness value (based on its objective function value) and fitter individuals are given a higher chance to mate and yield more “fitter” individuals. This is in line with the Darwin Theory of “Survival of the Fittest”. In this way we keep better individuals or solutions over generations, till we reach stopping criterion.
Note: Genetic algorithm are sufficiently randomized in nature, but they perform much better than random local search (In which we try just random solutions, keeping track of best so far), as they exploit historical information as well.

###Advantages of Genetic Algorithms:
1.   Is faster and more efficient as compared to the traditional methods.
2.   Has very good parallel capabilities.
3.   Optimize both continuous and discrete functions and also multi objective problems.
4.   Provides a list of good solution and not just a single solution.
5.   Always get answer to problem, which gets better over time.
6.   Useful when search space is very large and there are large number of parameters involved.
7.   Does not require any derivative information (Which may not be available for many real-world problems)

###Disadvantages of Genetic Algorithms:
1.   Fitness value is calculated repeatedly which might be computationally expensive for some problems.
2.   Being stochastic there is no guarantees on the optimality or quality of the solution.
3.   If not implemented properly, the GA may not converge to optimal solution.

###Basic Terminology:

* **Population:** It is the subset of all possible solutions to given problem. The population for GA is analogous to the population for human beings except that instead of human beings except that instead of human beings, we have candidate solutions representing human beings.
*	**Chromosomes**: A chromosome is one such solution to given problem.
*	**Gene**:  A gene is one element position of a chromosome.
*	**Allele**: It is the value a gene take for a particular chromosome.
*	**Genotype**: Genotype is solution in the computational space. In the computation space, the solutions are represented in a way which can be easily understood and manipulated using a computing system.
*	**Phenotype**: Phenotype is the population in the actual real-world solution space in which the solutions are represented in a way they are represented in real world situations.
*	**Decoding and Encoding**: For simple problems, genotype and phenotype spaces are same. However in most cases the, genotype and phenotype spaces are different e.g. 0/1 Knapsack problem. Decoding is the process of transforming a solution from genotype to the phenotype space, while encoding is the process of transforming from phenotype to genotype space. Decoding should be fast as it is carried out repeatedly in a GA during fitness value calculation.
*	**Fitness Function**: A fitness function is defined as a function which takes the solution as input and produces the suitability of the solution as the output. In some cases, the fitness function and the objective function may be the same, while in others it might be different based on the problem.
*	**Genetic Operators**: These alter the genetic composition of the offspring.  These include crossover, mutation, selection, etc.

### Basic Structure:
The basic structure of GA is as follows:

We start with initial population (which may be generated at random), select parents from this population for mating. Apply crossover and mutation operators on the parents to generate new off springs. And finally these off-springs replace the existing individuals in the population and process repeats. In this way GA actually try to mimic the human evolution to some extent.
