## 9.3 Differentiation
Let us take some function ($y = f(x)$) and numerically differentiate it. We will be taking $\sin$ function as an example.


```python
import numpy as np
import matplotlib.pyplot as plt
 
def f(x):
  return(np.sin(x))
 
x = np.arange(0,10,0.01)
y = f(x)
 
plt.figure(figsize=(10,5))
plt.plot(x, y, 'b')
plt.show()
```

We will be using *numpy* *gradient* to compute the first devrative. 

```python
dy_dx = np.gradient(y, x)

plt.plot(x, y, label="Sin")
plt.plot(x, dy_dx, label="Diff.")
plt.legend(loc="best")
plt.show()
```
