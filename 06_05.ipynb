## 6.5 Reading the raster
In this section, we will read the data that we wrote in earlier section. We will read the raster data
that has noise in it.

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
import gdal
import numpy as np

file_name = "/content/gdrive/My Drive/Colab Notebooks/LearnPython/data/data.tif"

driver = gdal.GetDriverByName('GTiff')
dataset = gdal.Open(file_name, gdal.GA_ReadOnly)
geotransform = dataset.GetGeoTransform()
data = dataset.GetRasterBand(1).ReadAsArray()
dataset = None
```

Now we can visualize the data using *matplotlib*.

```python
import matplotlib.pyplot as plt

plt.matshow(data)
plt.colorbar()
plt.show()
```

We can also use *rasterio* library to read raster data. Let us try the sample example.

```python
import pip
import matplotlib.pyplot as plt

try:
    import rasterio
except ImportError:
  pip.main(['install', 'rasterio']) 
  import rasterio

r = rasterio.open(file_name)

print(r.meta)

plt.matshow(r.read(1))
plt.colorbar()
plt.show()
```
