## 7.9 Annotation
Apart from plotting lines, dot, legends etc., we may need to put additional information on the plot. This is called annotation. We can put different arrow, texts etc. to make our graph more clear. Below figure shows a plot having few arrows, texts to improve the readability of the graph.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,10)
y = 5*np.sin(t)


plt.plot(t, y, lw=3, color='m')
plt.xlabel("X")
plt.ylabel("Y")
plt.annotate("ridge", xy=(np.pi/2,5), xycoords='data', xytext=(-20,-75), 
             textcoords="offset points", arrowprops=dict(arrowstyle="->"))

plt.annotate("valley", xy=(1.5*np.pi,-5), xycoords='data', xytext=(-30,80), 
             textcoords="offset points", size=20,
             bbox=dict(boxstyle="round,pad=0.5", fc="0.8"),
             arrowprops=dict(arrowstyle="->"))

plt.annotate("Annotation", xy=(8,-5), xycoords='data', xytext=(-20,0), 
             textcoords="offset points", size=20,
             bbox=dict(boxstyle="round", fc="green"),
             fontsize=15)

plt.text(3.0, 0, "Down", color = "w", ha="center", va="center", rotation=90,
         size=15, bbox=dict(boxstyle="larrow,pad=0.3", fc="r", ec="r", lw=2))

plt.text(np.pi*2, 0, "UP", color = "w", ha="center", va="center", rotation=90, 
         size=15, bbox=dict(boxstyle="rarrow,pad=0.3", fc="g", ec="g", lw=2))

plt.show()
```
